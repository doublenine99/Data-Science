```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
#import the necessary module
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

train = pd.read_csv('./train.csv')

# create the Labelencoder object
le = preprocessing.LabelEncoder()
le_class = preprocessing.LabelEncoder()
le_class.fit_transform(train['class'])

for column in train.columns:
    train[column] =  le.fit_transform(train[column])
x = train.drop(['class','Id','veil-type'], axis = 1) #TODO: 
y = train['class']

X_train, X_test,Y_train,Y_test = train_test_split(x,y, test_size = 0.5)

#create object of the lassifier
neigh = KNeighborsClassifier(n_neighbors=1)
#Train the algorithm
neigh.fit(X_train, Y_train)
# predict the response
pred = neigh.predict(X_test)
# evaluate accuracy
print ("KNeighbors accuracy score : ",accuracy_score(Y_test, pred))

model = KNeighborsClassifier(n_neighbors=1)
model.fit(x,y)

test = pd.read_csv('./test.csv')



for column in test.columns:
    if(column != 'Id'):
        test[column] =  le.fit_transform(test[column])
x_test = test.drop(['Id','veil-type'], axis = 1)
y_predict = model.predict(x_test)
y_predict = le_class.inverse_transform(y_predict)
# print(y_predict)

outcols = [outcols for outcols in train.columns if outcols in ['Id', 'class']]
output = train[:2467][outcols]
output['Id'] = test['Id']
output['class'] = y_predict

print(output)

output.to_csv('submission.csv', index=False)
```
